```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
x = np.linspace(0, 5, 1000)
k = 0.00008636243
T = 500
Ef = 2.5

def fd(x):
  return (1) / (1 + np.e**((x - Ef)/(k*T)))

def mb(x):
  return np.e**(Ef/(k*T)) * np.e**(-x/(k*T))
```

```python
plt.plot(x,fd(x), label = "Fermi-Dirac")
plt.plot(x,mb(x), label = "Maxwell-Boltzmann")
plt.ylim(0,2)
plt.legend()
plt.show()
```

```python
value = 2.5
integral_fd = integrate.quad(fd,value,5)[0]
integral_mb = integrate.quad(mb,value,5)[0]

print("Integral between {} and {} of the Fermi-Dirac distribution = {}".format(value,5,integral_fd))
print("Integral between {} and {} of the Maxwe-Boltz distribution = {}".format(value,5,integral_mb))
print("Ratio = {}".format(integral_fd/integral_mb))
```

```python
r = []
flag = 0
threshold = 0.9
for i in x[0:-1]:
  r.append(integrate.quad(fd,i,5)[0] / integrate.quad(mb,i,5)[0])
  if flag == 0 and r[-1] > threshold:
    flag = 1
    print("Valor mínimo de energía para superar un ratio de {} = {}".format(threshold,i))
```

```python
plt.title("Ratio entre Fermi-Dirac y Maxwell-Boltzmann")
plt.plot(x[0:-1],r)
plt.plot()
```
